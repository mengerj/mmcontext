```python
import anndata as ad
from adata_hf_datasets import InitialEmbedder
from adata_hf_datasets.dataset import AnnDataSetConstructor
from adata_hf_datasets.pp import preprocess_adata
```

```python
adata_path = "../../data/demo/demo.h5ad"
```

```python
adata = ad.read_h5ad(adata_path)
```

```python
adata.obs
```

```python
label_key = "cell_type"
batch_key = "dataset_title"
```

```python
# This will use raw counts layers, set them to X and apply log1p and normalisation.
adata_pp = preprocess_adata(adata, min_cells=1, min_genes=1, batch_key=batch_key)
```

```python
from adata_hf_datasets.pp.pybiomart_utils import ensure_ensembl_index

# might have to rename with your dataset.
# If your dataset does not have ensembl ids, use gene names, set them as the index, and the method will fetch the ensembl ids from biomaRt.
ensure_ensembl_index(adata_pp, ensembl_col="ensembl_id")
```

```python
embedder = InitialEmbedder(method="gs10k")
embedder.prepare(adata_pp)
emb_matrix = embedder.embed(adata_pp)
```

```python
adata_pp.obsm["X_gs10k"] = emb_matrix
```

```python
# write to file
adata_pp_path = "../../data/demo/demo_pp.h5ad"
adata_pp.write_h5ad(adata_pp_path)
```

```python
constructor = AnnDataSetConstructor()
constructor.add_anndata(
    adata=adata_pp, sentence_keys=["sample_index"], caption_key="natural_language_annotation", adata_link=adata_pp_path
)
ds = constructor.get_dataset()
# you can create several datasets for differennt splits and join them with DatasetDict
```

```python
# load the model you want to evaluate. It should be trained on the same initial embeddings as you applied to the inference data
from sentence_transformers import SentenceTransformer

model_name = "jo-mengr/mmcontext-pubmedbert-gs10k-cxg_geo_bio_dim2048"
model = SentenceTransformer(model_name, trust_remote_code=True)
data_type = "gs10k"
layer_key = f"X_{data_type}"
text_only = False
primary_cell_sentence = "cell_sentence_1"  # set to cell_sentence_2 for text based models
```

```python
if not text_only:
    token_df, _ = model[0].get_initial_embeddings_from_adata_link(
        ds,
        layer_key=layer_key,
        axis="obs",  # if you want to use a layer from varm and tokenize gene names, set to var
    )
    model[0].register_initial_embeddings(token_df, data_origin=data_type)
```

```python
if not text_only:
    ds_ready = model[0].prefix_ds(ds, primary_cell_sentence)
```
