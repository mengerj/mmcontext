```python
from datasets import load_dataset
from sentence_transformers import SentenceTransformer, mine_hard_negatives

ds_name = "jo-mengr/descriptions_cell_types"
model_name = "NeuML/pubmedbert-base-embeddings"
ds = load_dataset(ds_name, split="train")

model = SentenceTransformer(model_name)

hard_negatives = mine_hard_negatives(ds, model, num_negatives=10)
```

```python
from datasets import DatasetDict

ds_new = DatasetDict({"train": hard_negatives})
ds_new.push_to_hub(
    ds_name, revision="hard_negatives", commit_message="add 10 hard negatives with NeuML/pubmedbert-base-embeddings"
)
```
